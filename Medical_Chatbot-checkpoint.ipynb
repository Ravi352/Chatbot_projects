```python
import openai

with open("HIDDEN.txt") as file:
    openai.api_key=file.read()

    from typing import Union
import openai

def get_api_response(prompt: str) -> Union[str, None]:
    try:
        response = openai.ChatCompletion.create(
          model="gpt-3.5-turbo-16k",
          messages=[
            {
              "role": "system",
              "content": "You are a helpful assistant."
            },
            {
              "role": "user",
              "content": prompt  # Include the user's message here
            }
          ],
          temperature=1,
          max_tokens=15000,
          top_p=0.9,
          frequency_penalty=0,
          presence_penalty=0
        )
        return response.choices[0].message["content"].strip()  # Extract the content from the message
    except Exception as e:
        print('ERROR', e)
        return None


def update_list(message: str, pl: list[str]) -> None:
    pl.append(message)

def create_prompt(message: str, pl: list[str]) -> str:
    p_message = f'\nHuman: {message}'
    update_list(p_message, pl)
    prompt = ''.join(pl)
    return prompt

def get_bot_response(message: str, pl: list[str]) -> str:
    prompt = create_prompt(message, pl)
    bot_response = get_api_response(prompt)

    if bot_response:
        # Check if the response starts with "\nAI: "
        if bot_response.startswith("\nAI: "):
            pos = bot_response.find('\nAI: ')
            bot_response = bot_response[pos + 5:]
        return bot_response
    else:
        return "Something went wrong..."

def main():
    while True:
        prompt_list = ["You will pretend to be a medical advice bot and only answer medical-related questions.",'\nHuman: Who is Taylor Swift?',
                   '\nAI: Sorry mate, I only answer medical-related questions,because I am a MEDICAL BOT ']
        user_input = input('You: ')

        # Check if user input message length exceeds token limit and truncate if necessary
        if user_input=="stop":
          break
        else:
          response = get_bot_response(user_input, prompt_list)
          print(f'Bot: {response}')

if __name__ == '__main__':
    main()
```
