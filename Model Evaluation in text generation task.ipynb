```python
import nltk

# Sample reference texts
reference_texts = ["The quick brown fox jumps over the lazy dog.",
                   "A fast brown fox leaps over a sleepy dog."]

# Sample generated text
generated_text = "A quick brown fox jumps over the lazy dog."

# Tokenize the reference texts and generated text
reference_tokenized = [nltk.word_tokenize(ref) for ref in reference_texts]
generated_tokenized = nltk.word_tokenize(generated_text)

# Calculate BLEU score
bleu_score = nltk.translate.bleu_score.sentence_bleu(reference_tokenized, generated_tokenized)

print("BLEU Score:", bleu_score)
```
