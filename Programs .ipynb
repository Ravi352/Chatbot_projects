```python
## program Number Three
## Perfect number
## Concept: sum of its divisors will be equal to its digit 

def perfect_num(n):
  temp=n
  sum=0
  for i in range (1,n):
    if (n%i==0):
      sum=sum+i

## checking for the number to be perfect or not
  if (temp==sum):
    print("the number is perfect number")

  else:
    print("the number is not the perfect number")

perfect_num(8)
```

```python
## Fourth program 
## Strong number
## Concept: sum of the factorial of each digit of the number is equal to the number
## making factorial function using reccursion

def fact(n):
  if n==1:
    return 1
  else:
    return (n*fact(n-1))

def strong_number(n):
  temp=n
  sum=0
  while (n>0):
    dig=n%10
    sum=sum+fact(dig)
    n = n//10
  if (temp==sum):
    print("the number is strong number")
  else:
    print("the number is not the strong number")

strong_number(145)
```

```python
## Angstrom number 
## concept: sum of the cubes of the digit of the number equal to itself

def angstrom_number(n):
  temp=n
  sum=0
  while (n>0):
    dig=n%10
    sum=sum+dig**3
    n = n//10
  if (temp==sum):
    print("the number is angstrom number")
  else:
    print("the number is not the angstrom number")

angstrom_number(371)
```

```python
## prime number using recurssion

# Python 3 Program to find whether
# a Number is Prime or Not using
# Recursion

# Returns true if n is prime, else
# return false.
# i is current divisor to check.
def isPrime(n, i = 2):

	# Base cases
	if (n <= 2):
		return True if(n == 2) else False
	if (n % i == 0):
		return False
	if (i * i > n):
		return True

	# Check for next divisor
	return isPrime(n, i + 1)


# Driver Program
n = 15
if (isPrime(n)):
	print("Yes")
else:
	print("No")
	
# This code is contributed by
# Smitha Dinesh Semwal
```

```python
## factorial using loop
def factorial_loop(n):
  fact=1
  if (n<0):
    print("factorial cant be found out")
  elif (n==1):
    print(n)
  else:
    for i in range (1, n+1):
      fact=fact*i
  print(fact)
factorial_loop(6)
```

```python
def CheckPrime(i,num):
    if num==i:
        return 0
    else:
        if(num%i==0):
            return 1
        else:
            return CheckPrime(i+1,num)
num=int(input("Enter your Number:"))
if(CheckPrime(2,num)==0):
    print("It is a Prime Number.")
else:
    print("It is not a Prime Number.")
```
